> # **Introduction to Epochs, What are they??**

In machine learning, an epoch refers to **one complete pass through the entire training dataset during the training of a model** During each epoch, the model's parameters are adjusted based on the optimization algorithm and the computed error (or loss) between the predicted output and the actual target values.

> **Training Process**

The training process is typically divided into multiple epochs to ensure that the model learns from the entire dataset. 

> **What each Epoch consists of?**

Each epoch consists of **forward propagation**, where input data is passed through the model to generate predictions, and **backward propagation**, where the model's parameters are updated based on the computed loss. This iterative process helps the model improve its performance over time by adjusting its weights and biases.

The number of epochs is a hyperparameter that is set before the training process begins. Choosing an appropriate number of epochs is crucial, as too few may result in an underfit model that hasn't learned the underlying patterns, while too many may lead to overfitting, where the model becomes too specialized to the training data and performs poorly on new, unseen data.

> **Why do we normalize??**

normalizing features in a dataset is a common practice in machine learning to address issues related to scale differences, improve convergence speed, and enhance the generalization capability of models.

```python
# Now we need to normalize the pixel values of the training images and test images

# eleventh step
# normalization
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
# twelveth step
# build a classification model
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

> **Explanation of the above code**

In the code, we shaped a 'model' using the 'Sequential' function. Adding flair, we applied 'ReLU' for energy and 'Softmax' for finesse in the activation functions. It's a dance of transformations, creating a harmonious neural network.

> **The softmax function**

The softmax function is like a translator for numbers, especially in tasks like classifying things. It takes a bunch of numbers (scores) and turns them into probabilities, showing the chances of each option being the right one. The highest probability usually indicates the model's best guess for the correct answer.

> **The ReLU function and the types of ReLU**

ReLU, or Rectified Linear Unit, is a popular activation function in neural networks. It's straightforward and effective. Here are a few types and a brief explanation:

**Regular ReLU:**

* If the input is positive, it keeps it as is.
* If the input is negative, it turns it into zero.

**Leaky ReLU:**

* Similar to regular ReLU but allows a small, positive slope for negative inputs, preventing complete elimination.

**Parametric ReLU (PReLU):**

* Like Leaky ReLU but the slope becomes a learnable parameter during training.
* ReLU is chosen for its simplicity and ability to introduce non-linearity, making it useful for learning complex patterns in data.

```python
# thirteenth step
# compiling our model

model.compile(optimizer = tf.optimizers.Adam(),
              loss= 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

# fourteenth step 
# training our model by the model.fit() function
model.fit(training_images, training_labels, epochs=5)
```

> **Explanation of the code above (compiling our model)**

we compiled our model and used the **adam optimizer**. than in the loss, we wrote **sparse_categorical_crossentropy** which basically refers to a specific loss function commonly used in machine learning, especially in the context of neural network training for classification tasks.

```python
# fifteenth step
# evaluate our model with the unseen data
model.evaluate(test_images, test_labels)
```

> **The Callback function**

The callback function is a versatile tool that serves multiple purposes in training a machine learning model. If your goal is to halt the training process as soon as certain accuracy and loss thresholds are achieved, leveraging the callback function becomes essential. It allows you to incorporate specific conditions during training, ensuring your model stops when desired performance metrics are met.

```python
import tensorflow as tf

class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        '''
        Halts the training after reaching a certain loss threshold.

        Args:
        epoch (integer) - index of epoch (required but unused in the function definition below)
        logs (dict) - metric results from the training epoch
        '''

        # Check the loss
        if logs.get('loss') is not None and logs.get('loss') < 0.25:
            # Stop training if the loss is below the threshold
            print("\nLoss is lower than 0.25, so cancelling training!")
            self.model.stop_training = True

# Instantiate the class
callbacks = MyCallback()
```
